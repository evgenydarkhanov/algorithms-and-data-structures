**Задача**: наивный алгоритм построения суффиксного дерева для строки из латинского алфавита
- препроцессинг текста за $O(n^2)$
- поиск паттернов за $O(n)$


1. изначально `root` пустой
2. выписываем все суффиксы строки
3. есть ли у `root` дуга $i$-го суффикса?
    - нет --> создаём
    - да --> идём по $\exists$ дуге и делаем внутреннюю вершину
4. повторяем со $i + 1$ суффиксом
