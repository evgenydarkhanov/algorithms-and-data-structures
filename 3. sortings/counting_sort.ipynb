```python
import random
```

```python
def counting_sort(arr):
    max_value = max(arr)
    len_arr = len(arr)

    tmp = [0 for _ in range(max_value + 1)]
    out = [0 for _ in range(len_arr)]

    # заполняем tmp
    for elem in arr:
        tmp[elem] += 1

    # считаем префиксные суммы
    for i in range(1, max_value + 1):
        tmp[i] += tmp[i-1]

    # заполняем выходной массив справа налево
    for i in range(len_arr-1, -1, -1):
        out[tmp[arr[i]] - 1] = arr[i]
        tmp[arr[i]] -= 1

    return out
```

```python
def test_array_counting_sort(max_len, max_value):
    arr = [random.randint(0, max_value) for _ in range(max_len)]
    return arr
```

```python
arr = test_array_counting_sort(1000, 100)
print(arr[:20])
print(counting_sort(arr))
```
